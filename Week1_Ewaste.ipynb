```python
import os
import matplotlib.pyplot as plt

# For image visualization and operations (future use)
import cv2
import numpy as np

# Define the dataset path
base_path = 'modified-dataset'
train_path = os.path.join(base_path, 'train')
test_path = os.path.join(base_path, 'test')

def count_images(folder_path):
    classes = os.listdir(folder_path)
    image_count = {}

    for category in classes:
        class_folder = os.path.join(folder_path, category)
        if os.path.isdir(class_folder):
            image_count[category] = len(os.listdir(class_folder))

    return image_count

train_counts = count_images(train_path)
test_counts = count_images(test_path)

print("Training Set:", train_counts)
print("Testing Set:", test_counts)


plt.figure(figsize=(8,5))
plt.bar(train_counts.keys(), train_counts.values(), color='skyblue')
plt.title("Training Set - Image Count per Category")
plt.xlabel("Category")
plt.ylabel("Number of Images")
plt.grid(axis='y')
plt.show()

plt.figure(figsize=(8,5))
plt.bar(test_counts.keys(), test_counts.values(), color='lightgreen')
plt.title("Testing Set - Image Count per Category")
plt.xlabel("Category")
plt.ylabel("Number of Images")
plt.grid(axis='y')
plt.show()


import random
from PIL import Image

sample_class = random.choice(os.listdir(train_path))
sample_img_path = os.path.join(train_path, sample_class, random.choice(os.listdir(os.path.join(train_path, sample_class))))
img = Image.open(sample_img_path)

print(f"Category: {sample_class}")
plt.imshow(img)
plt.axis('off')
plt.show()

```
